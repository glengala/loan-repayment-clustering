## Customer Repayment Patterns Analysis – Finance / Lending
#### Business Problem:

The company wanted to identify which customer groups are more likely to repay loans late, helping management understand repayment risk and inform lending strategies.

```python
#Load Data
import pandas as pd

df = pd.read_csv("../Data/LoanDataset.csv")
df.head()
print(f"Data shape: {df.shape}")
```

#### 1. Data Pre-Processing
##### 1.a. Removing Mising Data & understand data distribution

```python
# Check for missing value
total_na=df.isnull().sum()
print(f"\nMissing value: \n{total_na}")

perc=((total_na/len(df))*100).round(3)
print(f"\nPercentage of missing value(%): \n{perc}")

# Check for Duplicated Row
print(f"\nDuplicated Row: {df.duplicated().any()}")
print(f"\nDuplicated Row: {df.duplicated().sum()}")
print(f"\nDuplicated Row %: {((df.duplicated().sum()/len(df))*100).round(3)}")

# Check data type of each columns
print(f"\nDatatype of each column: \n{df.dtypes}")
# Drop Duplicated Row
# Since the duplicated row is 6/32586, which is appx 0.018%  
df=df.drop_duplicates()

# Drop missing value in columns that have low %
df=df.dropna(subset=["customer_id","employment_duration","loan_amnt","Current_loan_status"])

#Drop historical_default col due to high missing value 63.6%, we dont have extra information to fill on.
df=df.drop("historical_default",axis=1)

# Change categorical variables data type to numerical
def remove_unit(df,column_name) :
    t = []
    for i in df[column_name]:
        value = str(i).replace('£',"").replace(",","")
        t.append(value)
    return t

df["customer_income"]=remove_unit(df,"customer_income")
df["loan_amnt"]=remove_unit(df,"loan_amnt")
       
df["customer_income"]=df["customer_income"].astype("float64")
df["loan_amnt"]=pd.to_numeric(df["loan_amnt"])

print(df.describe())
```

Preliminary data exploration revealed extreme and implausible values in customer_age and income. Outlier detection led to removal of invalid records and capping of extreme income values at the 99th percentile, ensuring the analysis reflected realistic borrower behaviour.

Outlier validation was performed using both statistical thresholds and domain context to distinguish genuine extremes from entry errors.

```python
import numpy as np
# Remove outliers:
# Remove customer age >100

df=df[df["customer_age"]< 100]

# Capping customer income to 99th percentile
per=np.percentile(df["customer_income"],99)
#print(per)

df=df[df["customer_income"]<=per]

# Loan amount adjustment easier to plot
df["loan_amnt"]=df["loan_amnt"]/1000
print(df.describe())
df=df.rename(columns={"loan_amnt":"loan_amount_in_thousand"})

# Capping loan amount to 99th percentile (due to large variance and outlier in data
per_loan=np.percentile(df["loan_amount_in_thousand"],99)
df=df[df["loan_amount_in_thousand"]<=per_loan]


# Noted that max loan interest rate is 23.22% while the mean is 11%. Need to be aware.
```

```python
# Imputing missing values for loan_int_rate based on the loan intention, loan grade and term year.
# Since the loan_int_rate has extreme outlier as noted perviously, we will use median value to impute the mising values

df.head(5)

df["loan_int_rate"]=df.groupby(["loan_intent","loan_grade","term_years"])["loan_int_rate"].transform(lambda x: x.fillna(x.median()))
df["loan_int_rate"].isna().sum()
#print(df["loan_int_rate"].isna())
# For group with no observed interest rate, median rates from adjacent term lengths were used as a proxy.
# Reflecting expected smooth variation in rates with loan term.

filter=df[(df["loan_intent"]=="PERSONAL")&(df["loan_grade"]=="E")&(df["term_years"]<=7)&(df["term_years"]>=5)]

median=filter["loan_int_rate"].median(skipna=True)

df["loan_int_rate"]=df["loan_int_rate"].fillna(median)
df["loan_int_rate"].isna().sum()
```

```python
# Plot data distribution
import matplotlib.pyplot as plt

def his (df, column_name):
    plt.hist(df[column_name],bins=50)
    plt.xlabel(column_name)
    plt.ylabel("Frequency")
    plt.title(f"{column_name} Distribution")
    plt.show()

his(df,"customer_age")
his(df,"customer_income")
his(df,"loan_amount_in_thousand")
```

##### 1.b. Handling Categorical Variables for downstream analysis

```python
# check current data type for each column
print(f"\nDatatype of each column: \n{df.dtypes} \n" )

# check value count for the categorical variables
print("_"*70,"\n")
print(df.home_ownership.value_counts(),"\n")

print(df.loan_intent.value_counts(),"\n")

print(df.loan_grade.value_counts(),"\n")

print(df.Current_loan_status.value_counts(),"\n")

# Deal with Ordinal Variables
df["loan_grade"] = df["loan_grade"].replace({"A": 1, "B": 2, "C": 3, "D":4, "E":5})
df["Current_loan_status"] = df["Current_loan_status"].replace({"NO DEFAULT": 0, "DEFAULT": 1})

# Deal with nominal variables
df = pd.get_dummies(df, columns=["home_ownership","loan_intent"])
print(df.head())

# recheck current data type for each column
print(f"\nDatatype of each column (after preprocessed): \n{df.dtypes} \n" )
```

#### 2. Exploratory Data Analysis (EDA)

When assessing default risk I focus on the following interpretable predictors first (*pattern spotting*): 
- *loan_grade*
- *loan_amount*
- *income*
  
*historical_default* would be useful but is 63.6% missing and not reliably imputable, so I removed it and use income as a proxy for repayment capacity. 

Decisions and sensitivity checks are documented before modelling.

```python
# Define Income Groups using quantile-cut for easy interpretation
df["Income_bracket"]=pd.qcut(df.customer_income, q=5, labels=["very low", "low","medium","high","very high"])
print(df["Income_bracket"].head(5))

# Define Loan Amount Groups using quantile-cut
df["Loan_amount_bracket"]=pd.qcut(df.loan_amount_in_thousand, q=5, labels=["very low", "low","medium","high","very high"])
print(df["Loan_amount_bracket"].head(5))
```

```python
# Default Rate by Income Bracket
income=df.groupby("Income_bracket",observed=False)["Current_loan_status"].mean()*100
print(income)
income.plot(kind='bar',x="Income Bracket", y="Default Rate in %", color='skyblue')
plt.title("Default Rate by Income Bracket")
plt.show()

# Default Rate by Loan Amount Bracket
loan_amount=df.groupby("Loan_amount_bracket",observed=False)["Current_loan_status"].mean()*100
print(loan_amount)
loan_amount.plot(kind='bar',x="Loan Amount Bracket", y="Default Rate in %", color='lightpink')
plt.title("Default Rate by Loan Amount")
plt.show()


# Default Rate by Loan Grade
loan_grade=df.groupby("loan_grade",observed=False)["Current_loan_status"].mean()*100
print(loan_grade)
loan_grade.plot(kind='bar',x="Loan Grade", y="Default Rate in %", color='gray')
plt.title("Default Rate by Loan Grade")
plt.show()
```

```python
# Check total loan distribution in each group to understand the default rate
    
freq_income=df["Income_bracket"].value_counts().sort_values()
print(freq_income)
his(df,"Income_bracket")

freq_loan_amount=df["Loan_amount_bracket"].value_counts().sort_values()
print(freq_loan_amount)
his(df,"Loan_amount_bracket")

freq_loan_grade=df["loan_grade"].value_counts().sort_values()
print(freq_loan_grade)
his(df,"loan_grade")
```

*NOTE:* The income and loan amount groups look healthy; no red flags there. But the loan grade imbalance matters. When one category (like grade 1) dominates, the overall patterns might mostly reflect that group’s behavior. The tiny grade 5 sample — 288 loans — explains why its 72% default rate looks extreme; with such a small base, that percentage can swing wildly.

#### 3. Data Analysis:
##### 3.a. Correlation and feature relationships

```python
import seaborn as sns

# prep df for correlation matrix
df1=df.drop(["customer_id","Income_bracket","Loan_amount_bracket"],axis=1)
bool_col=df1.select_dtypes(include="bool").columns.tolist()
df1[bool_col]=df1[bool_col].astype(int)

# create correlation matrix using spearman
res=df1.corr(method="spearman") # mitigates the effect of outliers and skewed distributions which lacked in Pearson’s method
print(res)

# create correlation matrix using kendall (use as cross check tool with spearman method)
res_k=df1.corr(method="kendall") # mitigates the effect of outliers and skewed distributions which lacked in Pearson’s method
print(res_k)
```

```python
# visualise correlation matrix - spearman
plt.figure(figsize=(20,20))
sns.heatmap(res,annot=True,cmap="coolwarm")
plt.show()
```

```python
# visualise correlation matrix - Kendall method
plt.figure(figsize=(20,20))
sns.heatmap(res_k,annot=True,cmap="coolwarm")
plt.show()
```

*Note:* Both methos highlight 5 variables that highly correlated to current_loan_status. These are *customer_income*,*loan_grade*, *loan_int_rate*, *home_ownership_Mortgage*,*home_ownership_Rent*.  Therefore, we will focus on these variables in our modelling steps.

```python
# Hypothesis testing for home_ownership_Mortgage and Current_loan_status
# Using Chi-square because both variables are categorical:
# Current_loan_status (0 = repaid on time, 1 = late/defaulted)
# home_ownership_MORTGAGE (False = no, True = yes)

from scipy.stats import chi2_contingency

cont_table=pd.crosstab(df["home_ownership_MORTGAGE"],df["Current_loan_status"])
stat, p, dof, expected = chi2_contingency(cont_table)

# interpret p-value
alpha=0.05
print(f"p-value is {p}")
if p<=alpha:
    print("Reject H0")
else:
    print("Fail to reject H0")

#p-value <0.05, we fail to reject H0, therefore, there is statistically significant correlation between 2 variables
```

```python
# Hypothesis testing for home_ownership_Mortgage and Current_loan_status

from scipy.stats import chi2_contingency

cont_table=pd.crosstab(df["home_ownership_RENT"],df["Current_loan_status"])
stat, p, dof, expected = chi2_contingency(cont_table)

# interpret p-value
alpha=0.05
print(f"p-value is {p}")
if p<=alpha:
    print("Reject H0")
else:
    print("Fail to reject H0")

#p-value <0.05, we fail to reject H0, therefore, there is statistically significant correlation between 2 variables
```

##### 3.b. Clustering Analysis

```python
# K-Mode: https://www.geeksforgeeks.org/machine-learning/k-mode-clustering-in-python/
#pip install kmodes
from kmodes.kmodes import KModes

# Create loan interest bracket
df["loan_int_rate_bracket"]=pd.qcut(df.loan_int_rate, q=5, labels=["very low rate", "low rate","medium rate","high rate","very high rate"])
#print(df["loan_int_rate"].head(5))

df["loan_grade"] = df["loan_grade"].replace({1:"A", 2:"B", 3:"C", 4:"D", 5:"E"})
# Reconstruct a single categorical column ready for k-modes

def combine_home_ownership(row):
    if row["home_ownership_MORTGAGE"]:
        return "MORTGAGE"
    elif row["home_ownership_RENT"]:
        return "RENT"
    else:
        return "OTHER"

df["home_ownership"] = df.apply(combine_home_ownership, axis=1)
df["home_ownership"] = df["home_ownership"].astype("category")
#print(df.dtypes)

X=df[["Income_bracket","loan_grade", "loan_int_rate_bracket", "home_ownership"]]
# Find the number of cluster using Elbow Method

cost = []
K = range(1,8)
for k in list(K):
    kmode = KModes(n_clusters=k, init = "random", n_init = 5, verbose=1)
    kmode.fit_predict(X)
    cost.append(kmode.cost_)
    
# Plot the Elbow Method to decide on the best "k"   
plt.plot(K, cost, 'x-')
plt.xlabel('No. of clusters')
plt.ylabel('Cost')
plt.title('Elbow Curve')
plt.show()
```

```python
# Clustering using k=4 from the above result

kmode=KModes(n_clusters=4, init="random",n_init=5,verbose=1)
df["Cluster"]=kmode.fit_predict(X)

# Profile each cluster based on the most commin (mode) value in each cluster
cluster=(df.groupby("Cluster")[["Income_bracket","loan_grade", "loan_int_rate_bracket", "home_ownership"]].agg(lambda x: x.value_counts().index[0]))
display(cluster)
df.to_csv("../Results/Cluster_profile.csv", index=False)
# Cross-Tabulate showing the % of default loan
late_rate=pd.crosstab(df["Cluster"], df["Current_loan_status"], normalize="index") * 100
print(late_rate)

df.to_csv("../Results/Cluster_defautl_rate.csv", index=False)
```

```python
# Check if the differences are real
from scipy.stats import chi2_contingency

table = pd.crosstab(df["Cluster"], df["Current_loan_status"])
chi2, p, dof, expected = chi2_contingency(table)

# interpret p-value
alpha=0.05
print(f"p-value is {p}")
if p<=alpha:
    print("Reject H0")
else:
    print("Fail to reject H0")

#p-value <0.05, we fail to reject H0, therefore, repayment rates differ significantly across clusters
```

### Conclusion:

Using K-Modes clustering on categorical features (income bracket, loan grade, interest-rate bracket, and home ownership), four distinct customer groups were identified.
Repayment outcomes differ significantly across these clusters.
Cluster 3, consisting primarily of low-income renters with high-interest loans (Grade C), shows the highest late repayment rate (≈ 50.7%).
Conversely, Cluster 2, made up of high-income mortgage holders with low-interest loans (Grade A), has the lowest rate (≈ 8.3%).
This suggests that income level, loan cost, and housing status jointly drive repayment risk. Targeted risk mitigation or stricter loan criteria should focus on Cluster 3 profiles.
